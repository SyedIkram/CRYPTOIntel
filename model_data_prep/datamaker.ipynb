```python
import pandas as pd
from textblob import TextBlob
import datetime
```

```python
newsDF = pd.read_csv('NewsData.csv')
newsDF.head(5)
```

```python
newsDF = newsDF[newsDF['tags'].str.lower().contains("xrp") | newsDF['tags'].str.lower().contains("ripple")\
                      | newsDF['categories'].str.lower().contains("xrp") |newsDF['categories'].str.lower().contains("ripple") |\
                     newsDF['body'].lower().contains("xrp") | newsDF['body'].lower().contains("ripple") ]
newsDF['sentiment'] = newsDF['body'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

```python
newsDFnew = newsDF[['date','sentiment']]
```

```python
newsDFnew.tail(5)
```

```python
newsDFnew = newsDFnew[newsDFnew['date'].str.contains("2018") | newsDFnew['date'].str.contains("2019")]
newsDFnew = newsDFnew.groupby(['date']).sentiment.agg('sum')
# newsDFnew = df[df['A'].isin([3, 6])]
newsDFnew.reset_index().head(5)
```

```python
newsDFnew.to_csv('sentimentBTCnews.csv',index=False)
```
